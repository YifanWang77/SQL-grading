# Preprocess

```python
import pandas as pd
```

```python
df = pd.read_csv('question&answer.csv')
```

```python
df
```

```python
df['preamble'][12895]
```

```python
import re

# Function to clean text
def clean_text(text):
    start_pos = re.search('Question:', text).start()
    return text[start_pos:]

# Apply the cleaning function to all rows in the 'preamble' column
df['cleaned_preamble'] = df['preamble'].apply(clean_text)
```

```python
df
```

```python
df = df.drop(columns=['preamble'])
```

```python
df = df.loc[:,['cleaned_preamble', 'submitted_answer', 'is_correct']]
```

```python
df
```

```python
# Function to add content to each row
def add_content(row):
    return "[Answer]: " + str(row)

# Apply the function to the 'cleaned_preamble' column
df['submitted_answer'] = df['submitted_answer'].apply(add_content)
```

```python
df
```

```python
# Function to add square brackets around "Question"
def add_brackets(row):
    return row.replace("Question", "[Question]")

# Apply the function to the 'cleaned_preamble' column
df['cleaned_preamble'] = df['cleaned_preamble'].apply(add_brackets)
```

```python
df
```

```python
df['text'] = df['cleaned_preamble'] + " " + df['submitted_answer']

df
```

```python
df = df.drop(columns='cleaned_preamble').drop(columns='submitted_answer')
```

```python
df = df.loc[:,['text', 'is_correct']]
```

```python
df
```

```python
df = df.rename({'is_correct': 'label'}, axis=1)
```

```python
df
```

```python
df['text'][0]
```

```python
df.info()
```

# Training

```python
import numpy as np
import pandas as pd
from datasets import Dataset
from transformers import TrainingArguments
from transformers import Trainer
```

```python
ds = Dataset.from_pandas(df)
```

```python
ds
```

```python
ds = ds.train_test_split(test_size=0.2)
ds
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
```

```python
tokenizer(ds['train'][0]['text'])
```

```python
def tokenize(batch):
    return tokenizer(batch['text'], truncation=True)
```

```python
tokenize(ds['train'][:3])
```

```python
tokenized_ds = ds.map(tokenize, batched=True)
tokenized_ds
```

```python
from transformers import DataCollatorWithPadding

data_collator = DataCollatorWithPadding(tokenizer)
```

```python
import evaluate

accuracy = evaluate.load("accuracy")

def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    predictions = np.argmax(predictions, axis=1)
    return accuracy.compute(predictions=predictions, references=labels)
```

```python
id2label = {0: "negative", 1: "positive"}  
label2id = {"negative": 0, "positive": 1}
```

```python
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained(
    "bert-base-uncased", num_labels=2, id2label=id2label, label2id=label2id
)
```

```python
from transformers import TrainingArguments

finetuned_model_name = 'PatWang/bert-base-uncased-finetuned-sql-classification-with_question'
batch_size = 16

training_args = TrainingArguments(
    output_dir=finetuned_model_name,
    num_train_epochs=10,
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size,
    evaluation_strategy='epoch',
    save_strategy='epoch'
)
```

```python
from transformers import Trainer

trainer = Trainer(
    model,
    training_args,
    train_dataset=tokenized_ds['train'],
    eval_dataset=tokenized_ds['test'],
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)
```

```python
preds_output = trainer.predict(tokenized_ds['test'])
preds_output.metrics
```

```python
trainer.train()
```

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-classification", model="PatWang/bert-base-uncased-finetuned-sql-classification-with_question")
```

```python
df['text'][0]
```

```python
df_test = df.iloc[:1000]
```

```python
predictions = []

for i in range(1000):
    text = df_test.iloc[i]['text']
    pred = pipe(text)
    predictions.append(pred)

df_test['pred'] = predictions
```

```python
df_test['predicted_label'] = df_test['pred'].apply(lambda x: 1 if x[0]['label'] == 'positive' else 0)

df_test
```

```python

correct_predictions = (df_test['label'] == df_test['predicted_label']).sum()

accuracy = correct_predictions / len(df_test)

print(f"Accuracy: {accuracy}")
```
